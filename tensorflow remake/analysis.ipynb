```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
metrics = pd.read_csv('metrics.csv')
metrics.head()
```

```python
len(metrics)
```

```python
metric = 'CrossEntropyLoss'
```

```python
plt.plot(metrics[[metric, f'val_{metric}']], label=[metric, f'val_{metric}'])
plt.xlabel('Epoch')
plt.ylabel('Score')
plt.legend()
```

```python
metric = 'accuracy'
```

```python
plt.plot(metrics[[metric, f'val_{metric}']], label=[metric, f'val_{metric}'])
plt.xlabel('Epoch')
plt.ylabel('Score')
plt.legend()
```

```python
metric = "accuracy_class_"
```

```python
metric = [metric+str(i) for i in range(10)] 
plt.plot(metrics[metric], label=metric)
plt.xlabel('Epoch')
plt.ylabel('Score')
plt.legend()
```
